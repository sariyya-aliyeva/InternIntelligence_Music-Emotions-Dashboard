```python
import pandas as pd

data = pd.read_csv("SpotifyFeatures.csv")
```

```python
data.head()
```

```python
print(data.columns)
```

```python
data = data.dropna()
```

```python
data = data[['genre', 'artist_name', 'track_name', 'popularity',
             'danceability', 'energy', 'loudness', 'speechiness',
             'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
```

```python
print(data.info())
```

```python
valence_by_genre = data.groupby('genre')['valence'].mean().sort_values(ascending = False)
print(valence_by_genre)
```

```python
data.to_csv("spotify_clean.csv", index=False)
```
